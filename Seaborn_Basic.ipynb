```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

Relational plot

```python
# Scatter plot -> axis level
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# relplot -> figure level
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',style='time',size='size')
```

```python
# Line Plot
```

```python
gap = px.data.gapminder()
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# Axis level Function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# Using relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country')
```

```python
# Facet plot
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='smoker')
```

Distribution plots

```python
# figure level -> displot
# axes level -> hisplot -> kdeplot -> rugplot
```

```python
# plotting univarate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.histplot(data=tips,x='total_bill',bins=20)
```
